```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/spam.csv',encoding='ISO-8859-1')
data.rename(columns={'v1':'target','v2':'text'},inplace=True)
data.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['target']=encoder.fit_transform(data['target'])
data
```

```python
x=data['text']
y=data['target']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer(stop_words='english',max_features=5000)
xtrain_tf=vect.fit_transform(xtrain)
x_test_tf=vect.transform(xtest)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(xtrain_tf,ytrain)
y_pred_nb=model.predict(x_test_tf)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(ytest,y_pred_nb))
print(confusion_matrix(ytest,y_pred_nb))
print(classification_report(ytest,y_pred_nb))
```

```python
new_message={
    "Congratulations! You have won an iphone.Call me to claim your gift",
    "You've been selected for this job opportunity!Kindly call us for further details.",
    "Can we meet tomorrow?",
    "Hey!Are we still on for the meeting tomorrow?"
}
new_message_tf=vect.transform(new_message)
pred_nb=model.predict(new_message_tf)
label_map={0:'ham',1:'spam'}
for i,message in enumerate(new_message):
  print(f"Message:{message}")
  print(f"Naive Bayes Prediction:{label_map[pred_nb[i]]}")
```
