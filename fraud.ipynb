```python
if 'Unnamed: 0' in data.columns:
    data.drop('Unnamed: 0', axis=1, inplace=True)
if 'cc_num' in data.columns:
    data.drop('cc_num', axis=1, inplace=True)
if 'first' in data.columns:
    data.drop('first', axis=1, inplace=True)
if 'last' in data.columns:
    data.drop('last', axis=1, inplace=True)
if 'street' in data.columns:
    data.drop('street', axis=1, inplace=True)
if 'city' in data.columns:
    data.drop('city', axis=1, inplace=True)
if 'zip' in data.columns:
     data.drop('zip', axis=1, inplace=True)
if 'lat' in data.columns:
    data.drop('lat', axis=1, inplace=True)
if 'long' in data.columns:
    data.drop('long', axis=1, inplace=True)
if 'city_pop' in data.columns:
    data.drop('city_pop', axis=1, inplace=True)


non_numeric_cols = data.select_dtypes(include=['object']).columns
if len(non_numeric_cols) > 0:
    print(f"Warning: Non-numeric columns found: {list(non_numeric_cols)}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE

x=data.drop('is_fraud',axis=1)
y=data['is_fraud']
non_numeric_cols = x.select_dtypes(include=['object']).columns
if len(non_numeric_cols) > 0:
    print(f"Dropping non-numeric columns before SMOTE: {list(non_numeric_cols)}")
    x = x.drop(non_numeric_cols, axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
